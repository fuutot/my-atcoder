```python
# じゃんけん
n, k = map(int, input().split())
roc_score, scissors_score, paper_score = map(int, input().split())
T = input()
T = " " + T

papers = [0 for i in range(n+1)]
scissors = [0 for i in range(n+1)]
rocs = [0 for i in range(n+1)]

for i in range(1, k+1):
    now_hand = T[i]

    # rの時はpaper
    if now_hand == 'r':
        papers[i] = paper_score
    
    # pの時はscissors
    elif now_hand == 'p':
        scissors[i] = scissors_score

    # sの時はrocs
    else:
        rocs[i] = roc_score

for i in range(k+1, n+1):
    now_hand = T[i]

    # rの時はpaper
    if now_hand == 'r':
        if papers[i-k] == 0:
            papers[i] = paper_score
    
    # pの時はscissors
    elif now_hand == 'p':
        if scissors[i-k] == 0:
            scissors[i] = scissors_score

    # sの時はrocs
    else:
        if rocs[i-k] == 0:
            rocs[i] = roc_score

ans = sum(papers) + sum(scissors) + sum(rocs)
print(ans)
```

```python
# カード
n = int(input())
A = [None] * (n+1)
B = [None] * (n+1)
for i in range(1, n+1):
    A[i], B[i] = map(int, input().split())

From_A = [0] * (n+1)
From_B = [0] * (n+1)
From_A[1] = 1
From_B[1] = 1

for i in range(2, n+1):
    # 表と表が同じじゃないとき
    if A[i-1] != A[i]:
        From_A[i] += From_A[i-1] % 998244353
    # 表と裏が同じじゃないとき
    if A[i-1] != B[i]:
        From_B[i] += From_A[i-1] % 998244353
    # 裏と表が同じじゃないとき
    if B[i-1] != A[i]:
        From_A[i] += From_B[i-1] % 998244353
    # 裏と裏が同じじゃないとき
    if B[i-1] != B[i]:
        From_B[i] += From_B[i-1] % 998244353
    # 同じじゃないように選べない時
    elif A[i-1] == A[i] == B[i-1] == B[i]:
        break

print(From_A[-1] % 998244353 + From_B[-1] % 998244353)
```
