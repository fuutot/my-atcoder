```python
#a11
n, x = map(int, input().split())
A = list(map(int, input().split()))

def is_ok(arg):
    return A[arg] >= x

def binary_search(ng, ok):
    while abs(ok - ng) > 1:
        mid = (ok + ng) // 2
        if is_ok(mid):
            ok = mid
        else:
            ng = mid
    return ok

ans = binary_search(-1, n)
print(ans+1)
```

```python
#a12
n, k = map(int, input().split())
A = list(map(int, input().split()))

def is_ok(arg):
    total = 0
    for i in range(n):
        total += arg//A[i]
    return total >= k

def binary_search(ng, ok):
    while abs(ok - ng) > 1:
        mid = (ok + ng) // 2
        if is_ok(mid):
            ok = mid
        else:
            ng = mid
    return ok

print(binary_search(-1, 10**9+1))
```

```python
#a13
n, k = map(int, input().split())
A = list(map(int, input().split()))

def is_ok(arg):
    return A[arg] - origin <= k

def binary_search(ng, ok):
    while abs(ok - ng) > 1:
        mid = (ok + ng) // 2
        if is_ok(mid):
            ok = mid
        else:
            ng = mid
    return ok

cnt = 0
for i in range(n):
    origin = A[i]
    target_i = binary_search(n, i)
    cnt += target_i - i
print(cnt)
```

```python
#a14
n, k = map(int, input().split())
A = list(map(int, input().split()))
B = list(map(int, input().split()))
C = list(map(int, input().split()))
D = list(map(int, input().split()))

X = []
for i in range(n):
    for j in range(n):
        X.append(A[i] + B[j])

Y = []
for i in range(n):
    for j in range(n):
        Y.append(C[i] + D[j])
Y.sort()

def is_ok(arg):
    return Y[arg] + x >= k

def binary_search(ng, ok):
    while abs(ok - ng) > 1:
        mid = (ok + ng) // 2
        if is_ok(mid):
            ok = mid
        else:
            ng = mid
    return ok

ans = 'No'
for i in range(len(X)):
    x = X[i]
    y_i = binary_search(-1, len(Y))
    if y_i == len(Y):
        continue
    else:
        if x + Y[y_i] == k:
            ans = 'Yes'
            break
print(ans)
```
