```python
n, m = map(int, input().split())
A = [None for i in range(m)]
B = [None for i in range(m)]
for i in range(m):
    A[i], B[i] = input().split()

house = set()
ans = ["No" for i in range(m)]

for i in range(m):
    if A[i] not in house and B[i] == "M":
        ans[i] = "Yes"
        house.add(A[i])

for i in range(m):
    print(ans[i])
```

```python
n = int(input())
mg = int(input())
G = set()
for i in range(mg):
    u, v = map(int, input().split())
    G.add((u, v))
mh = int(input())
H = set()
for i in range(mh):
    a, b = map(int, input().split())
    H.add((a, b))
A = {k:[0]*9 for k in range(1, 9)}

for i in range(1,n):
    c = list(map(int, input().split()))
    for j in range(len(c)):
        A[i][i+j+1] = c[j]

cost = 0
for e in H:
    if e not in G:
        cost += A[e[0]][e[1]]

for e in G:
    if e not in H:
        cost += A[e[0]][e[1]]

print(cost)
```

```python
n = int(input())
X = list(map(int, input().split()))
P = list(map(int, input().split()))
q = int(input())
L = [None]*q
R = [None]*q
for i in range(q):
    L[i], R[i] = map(int, input().split())

Z = [0 for i in range(n+1)]

for i in range(n):
    Z[i+1] = Z[i] + P[i]

# めぐる式二分探索

def left_is_ok(arg, l):
    # 条件を満たすかどうか？問題ごとに定義
    return X[arg] >= l


def left_meguru_bisect(ng, ok, l):
    '''
    二分探索を行う関数。まず満たしてほしい条件をis_okに定義する
    params:
        ng: 満たさないindexを指定
        ok: 満たすindexを指定
    '''
    while (abs(ok - ng) > 1): # 隣あうまでやる
        mid = (ok + ng) // 2
        if left_is_ok(mid, l):
            ok = mid # okは常に条件を満たす
        else:
            ng = mid # ngは常に条件を満たさない
    return ok

# めぐる式二分探索

def right_is_ok(arg, r):
    # 条件を満たすかどうか？問題ごとに定義
    return X[arg] <= r


def right_meguru_bisect(ng, ok, r):
    '''
    二分探索を行う関数。まず満たしてほしい条件をis_okに定義する
    params:
        ng: 満たさないindexを指定
        ok: 満たすindexを指定
    '''
    while (abs(ok - ng) > 1): # 隣あうまでやる
        mid = (ok + ng) // 2
        if right_is_ok(mid, r):
            ok = mid # okは常に条件を満たす
        else:
            ng = mid # ngは常に条件を満たさない
    return ok

for i in range(q):
    l = L[i]
    r = R[i]
    l_i = left_meguru_bisect(-1, n, l)
    r_i = right_meguru_bisect(n, -1, r)
    if r_i < l_i:
        print(0)
    elif l_i == 0:
        print(Z[r_i+1])
    else:
        print(Z[r_i+1]-Z[l_i])
```
